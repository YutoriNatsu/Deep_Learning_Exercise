```python
def count_words_in_files(file_names):
    word_counts = []
    for file_name in file_names:
        try:
            with open(file_name, 'r', encoding='utf-8') as file:
                content = file.read()
                word_count = len(content)
                word_counts.append(word_count)
        except FileNotFoundError:
            print(f"File {file_name} not found.")
            word_counts.append(0)
        except Exception as e:
            print(f"An error occurred while reading {file_name}: {e}")
            word_counts.append(0)
    return word_counts

# Generate list of file names from 1.txt to 50.txt
file_names = [f"{i}.txt" for i in range(1, 51)]

# Count words in each file and store results in a list
word_counts = count_words_in_files(file_names)

print(word_counts)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成示例数据：5个样本，每个样本有6个特征
np.random.seed(0)
data = np.random.rand(5, 6) * 100  # 随机生成0到100之间的数值
columns = ['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5', 'Feature6']
samples = ['Sample1', 'Sample2', 'Sample3', 'Sample4', 'Sample5']

# 创建雷达图
num_vars = len(columns)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # 闭合图形

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# 绘制每个样本的数据
for i in range(len(samples)):
    values = data[i].tolist()
    values += values[:1]  # 闭合图形
    ax.plot(angles, values, label=samples[i])
    ax.fill(angles, values, alpha=0.25)

# 设置标签
ax.set_xticks(angles[:-1])
ax.set_xticklabels(columns)
ax.set_title('Radar Chart of 6D Data')

# 添加图例
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.show()
```
